```python
import glob
import os
import nltk
import pickle
from math import log
import re, string, unicodedata
import contractions
import inflect
from bs4 import BeautifulSoup
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.tokenize.punkt import PunktSentenceTokenizer
from collections import Counter
from random import randint
import re
nltk.download('punkt')
path = "C:\Users\MyPC\Desktop\Ass3\Ass3"
start_words = []
end_words = []

def create_sentence_tokens():
	print("creating sentence tokens...")
	os.chdir("C:\Users\MyPC\Desktop\Ass3\Ass3/20_newsgroups/comp.graphics/")
	all_sentence_tokens = []
# 	i = 0
	for file in glob.glob("*"):
# 		print i+1
# 		i+=1
		f = open(file, 'rb')
		sentences = PunktSentenceTokenizer().tokenize(f.read())
		all_sentence_tokens += sentences
		f.close()
	return all_sentence_tokens

def create_word_tokens():
	print("creating word tokens...")
	corpus = []
	for sentence in all_sentence_tokens:
		temp = []
		words = nltk.word_tokenize(sentence)
		for word in words:
			word = re.sub(r'\W+', '', word)
			word = re.sub(r'_+', '', word)
			if(word!=''):
				temp.append(word)
		if(len(temp)!=0):
			start_words.append(temp[0])
			end_words.append(temp[len(temp)-1])
			corpus.append(temp)
	return corpus


def find_log_probs_unigram():
	print("calculating probabilties...")
	log_probs_uni = {}
	total_count = 0
	for sentences in corpus:
		for words in sentences:
# 			print words
			log_probs_uni[words]=[log_probs_uni[words][0]+1] if words in log_probs_uni.keys() else [1]
			total_count += 1
	print("calculating log probabilties...")
	for word,val in log_probs_uni.iteritems():
		freq = val[0]
		prob = float(freq)/total_count
# # print corpus
# log_probs_uni = find_log_probs_unigram()t
		log_prog = log(prob)
		log_probs_uni[word].append(prob)
		log_probs_uni[word].append(log_prog)
		# break
# 	with open(path+"unigram_log_probs_1.txt", "wb") as myFile:
# 		pickle.dump(log_probs_uni, myFile)
	return log_probs_uni
	


all_sentence_tokens = create_sentence_tokens()
corpus = create_word_tokens()
# # print corpus
log_probs_uni = find_log_probs_unigram()
print log_probs_uni
```
